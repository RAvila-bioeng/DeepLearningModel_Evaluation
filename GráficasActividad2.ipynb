```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# === CONFIG ===
csv_folder = "/content/drive/MyDrive/unet_experimentos"  # <--- CAMBIA ESTO
output_folder = "/content/drive/MyDrive/gráfica"                    # donde guardo las gráficas

os.makedirs(output_folder, exist_ok=True)

# === PROCESAR TODOS LOS CSV ===
csv_files = [f for f in os.listdir(csv_folder) if f.endswith(".csv")]

for file in csv_files:
    path = os.path.join(csv_folder, file)
    print(f"Procesando: {file}")

    df = pd.read_csv(path)

    # ——— Datos del título ———
    param = df["parameter"][0]
    val = df["value"][0]

    # ——— Crear figura y ejes ———
fig, ax1 = plt.subplots(figsize=(10,6))
ax2 = ax1.twinx()

# ——— Gráficas de LOSS (eje izquierdo) ———
ax1.plot(df["epoch"], df["train_loss"], color="tab:red", linewidth=2, label="train_loss")
ax1.plot(df["epoch"], df["val_loss"], color="tab:red", linestyle="--", linewidth=2, label="val_loss")
ax1.set_ylabel("Loss", color="tab:red")
ax1.tick_params(axis='y', labelcolor="tab:red")

# ——— Gráficas de ACCURACY (eje derecho) ———
ax2.plot(df["epoch"], df["train_acc"], color="tab:blue", linewidth=2, label="train_acc")
ax2.plot(df["epoch"], df["val_acc"], color="tab:blue", linestyle="--", linewidth=2, label="val_acc")
ax2.set_ylabel("Accuracy", color="tab:blue")
ax2.tick_params(axis='y', labelcolor="tab:blue")

# ——— EJE X ———
ax1.set_xlabel("Epoch")
ax1.set_xticks(df["epoch"])  # muestra todos los números

# ——— Título ———
plt.title(f"{param} = {val}", fontsize=16)

# ——— LEYENDA SIEMPRE VISIBLE ———
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
fig.legend(
    lines1 + lines2,
    labels1 + labels2,
    loc="upper center",
    bbox_to_anchor=(0.5, 1),   # <--- acercamos la leyenda al gráfico
    ncol=4,
    fontsize=11,
    frameon=True
)

# ——— Ajuste profesional ———
plt.tight_layout(rect=[0, 0, 1, 0.95])  # <--- espacio reservado para la leyenda

# ——— Guardar figura ———
save_path = os.path.join(output_folder, f"{param}_{val}.png")
plt.savefig(save_path, dpi=300)
plt.close()

print("\nTodas las gráficas han sido generadas y guardadas en:", output_folder)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# === CONFIG ===
csv_folder = "/content/drive/MyDrive/unet_experimentos"  # ruta donde están tus CSV
output_folder = "/content/plots"                  # carpeta salida

os.makedirs(output_folder, exist_ok=True)

# === PROCESAR TODOS LOS CSV ===
csv_files = [f for f in os.listdir(csv_folder) if f.endswith(".csv")]

for file in csv_files:
    path = os.path.join(csv_folder, file)
    print(f"Procesando: {file}")

    df = pd.read_csv(path)

    # Datos para el título
    param = df["parameter"][0]
    val = df["value"][0]

    # ——— Crear figura y ejes ———
    fig, ax1 = plt.subplots(figsize=(10,6))
    ax2 = ax1.twinx()

    # ——— Gráficas de LOSS ———
    ax1.plot(df["epoch"], df["train_loss"], color="tab:red", linewidth=2, label="train_loss")
    ax1.plot(df["epoch"], df["val_loss"], color="tab:red", linestyle="--", linewidth=2, label="val_loss")
    ax1.set_ylabel("Loss", color="tab:red")
    ax1.tick_params(axis='y', labelcolor="tab:red")

    # ——— Gráficas de ACC ———
    ax2.plot(df["epoch"], df["train_acc"], color="tab:blue", linewidth=2, label="train_acc")
    ax2.plot(df["epoch"], df["val_acc"], color="tab:blue", linestyle="--", linewidth=2, label="val_acc")
    ax2.set_ylabel("Accuracy", color="tab:blue")
    ax2.tick_params(axis='y', labelcolor="tab:blue")

    # ——— EJE X ———
    ax1.set_xlabel("Epoch")
    ax1.set_xticks(df["epoch"])

    # ——— TÍTULO ———
    plt.title(f"{param} = {val}", fontsize=16, pad=20)

    # ——— LEYENDA ARRIBA ———
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    fig.legend(
        lines1 + lines2,
        labels1 + labels2,
        loc="upper center",
        bbox_to_anchor=(0.5, 1),
        ncol=4,
        fontsize=11,
        frameon=True
    )

    # ——— AJUSTE ———
    plt.tight_layout(rect=[0, 0, 1, 0.95])

    # ——— GUARDAR FIGURA ———
    save_path = os.path.join(output_folder, f"{param}_{val}.png")
    plt.savefig(save_path, dpi=300)
    plt.close()

print("\nTodas las gráficas han sido generadas y guardadas en:", output_folder)
```
